```python
puzzle = Puzzle(year=2019, day=3)
data = puzzle.input_data

wire_1 = data.split("\n")[0].split(",")
wire_2 = data.split("\n")[1].split(",")



def create_paths(wire):
    wire_path = []
    wire_direction_changes = [(0,0)]

    for i, element in enumerate(wire):
        direction = element[0]
        path = int(element[1::])
        starting_vert = wire_direction_changes[i][0]
        starting_horiz = wire_direction_changes[i][1]
        if direction == "R":
            for x in range(starting_horiz, starting_horiz + path ):
                wire_path.append((starting_vert, x))
            wire_direction_changes.append((starting_vert, starting_horiz + path))
        if direction == "L":
            for x in range(starting_horiz, starting_horiz - path, -1):
                wire_path.append((starting_vert, x))
            wire_direction_changes.append((starting_vert, starting_horiz - path))
        if direction == "U":
            for x in range(starting_vert, starting_vert + path):
                wire_path.append((x, starting_horiz))
            wire_direction_changes.append((starting_vert + path, starting_horiz))
        if direction == "D":
            for x in range(starting_vert, starting_vert - path, -1):
                wire_path.append((x, starting_horiz))
            wire_direction_changes.append((starting_vert - path, starting_horiz))
    return wire_path, wire_direction_changes

wire1_path, wire1_direction_changes = create_paths(wire_1)

wire2_path, wire2_direction_changes = create_paths(wire_2)

```

```python
minimum = 10000000000
for i,x in enumerate(wire1_path):
    for j,y in enumerate(wire2_path):
        if x == y:
            if i +j <= minimum: 
                minimum = i+j

print(minimum)
                
            
```

```python
wire1_path_set = set(wire1_path)
wire2_path_set = set(wire2_path)
intersection = wire1_path_set.intersection(wire2_path_set) 

print(intersection)

minimum = 1000000000
for x in intersection:
    if x != (0,0) and x != (0,1):
        dist = abs(x[0])+abs(x[1])
        if dist <= minimum: 
            minimum = dist
            point = x

print(minimum)
print(point)

steps = []
for x in intersection: 
    if x != (0,0):
        a = wire1_path.index(x)
        b = wire2_path.index(x)
        steps.append(a+b)
print(min(steps))
```

```python
wire1_ex = "R75,D30,R83,U83,L12,D49,R71,U7,L72".split(",")
wire2_ex = "U62,R66,U55,R34,D71,R55,D58,R83".split(",")

path1, dire1 = create_paths(wire1_ex)
path2, dire2 = create_paths(wire2_ex)

print(path1, dire1)

path1_set = set(path1)
path2_set = set(path2)
intersection = path1_set.intersection(path2_set)

print(intersection)

minimum = 10000
for x in intersection:
    if x != (0,0):
        dist = abs(x[0])+abs(x[1])
        if dist <= minimum: 
            minimum = dist
            point = x
print(point)
```
